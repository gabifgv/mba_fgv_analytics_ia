## Linear Regression Tutorial

This notebook demonstrates various concepts in linear regression, including:

Simple linear regression with one variable
Multiple linear regression with several variables
Handling non-linear features
Feature transformations and engineering
Model comparison and evaluation

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
import seaborn as sns
from matplotlib import style
style.use('ggplot')

# Set random seed for reproducibility
np.random.seed(42)
```

## PART 1: SIMPLE LINEAR REGRESSION (ONE VARIABLE)
First, we'll demonstrate simple linear regression with a single input variable.

```python
# Create a simple dataset for demonstration
n_samples = 100
X = np.linspace(0, 10, n_samples).reshape(-1, 1)
y = 2 * X.flatten() + 5 + np.random.normal(0, 2, n_samples)  # y = 2x + 5 + noise

# Visualize the data
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', alpha=0.5)
plt.title('Simple Linear Regression Dataset')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
# Create and train a simple linear regression model
model = LinearRegression()
model.fit(X, y)

# Get the model parameters
w = model.coef_[0]  # Slope (weight)
b = model.intercept_  # Intercept (bias)

# Make predictions
y_pred = model.predict(X)

# Calculate loss (mean squared error)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

# Plot the regression line
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', alpha=0.5, label='Data points')
plt.plot(X, y_pred, color='red', linewidth=2, label='Regression line')
plt.title('Simple Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()

print(f"Simple Linear Regression Results:")
print(f"Equation: y = {w:.4f}x + {b:.4f}")
print(f"True Equation: y = 2x + 5 + noise")
print(f"Mean Squared Error: {mse:.4f}")
print(f"R² Score: {r2:.4f}")
```

## PART 2: MULTIPLE LINEAR REGRESSION
Now, let's extend to multiple linear regression with several input variables.

```python
# Create a more complex dataset with multiple features
n_samples = 1000
X_multi = np.random.rand(n_samples, 3) * 10  # 3 features
# True relationship: y = 2*x1 + 0.5*x2 - 1.5*x3 + 5 + noise
y_multi = (2 * X_multi[:, 0] +
           0.5 * X_multi[:, 1]
           - 1.5 * X_multi[:, 2] +
           5 +
           np.random.normal(0, 2, n_samples))

# Create a DataFrame for better visualization
df_multi = pd.DataFrame(X_multi, columns=['feature1', 'feature2', 'feature3'])
df_multi['target'] = y_multi

# Visualize the relationships between features and target
fig, axs = plt.subplots(1, 3, figsize=(18, 5))
for i, feature in enumerate(['feature1', 'feature2', 'feature3']):
    axs[i].scatter(df_multi[feature], df_multi['target'], alpha=0.5)
    axs[i].set_title(f'{feature} vs target')
    axs[i].set_xlabel(feature)
    axs[i].set_ylabel('target')
plt.tight_layout()
plt.show()
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X_multi, y_multi, test_size=0.2, random_state=42
)

# Train multiple linear regression model
multi_model = LinearRegression()
multi_model.fit(X_train, y_train)

# Get model parameters
weights = multi_model.coef_
bias = multi_model.intercept_

# Make predictions
y_train_pred = multi_model.predict(X_train)
y_test_pred = multi_model.predict(X_test)

# Calculate metrics
train_mse = mean_squared_error(y_train, y_train_pred)
test_mse = mean_squared_error(y_test, y_test_pred)
train_r2 = r2_score(y_train, y_train_pred)
test_r2 = r2_score(y_test, y_test_pred)

print(f"Multiple Linear Regression Results:")
print(f"Equation: y = {weights[0]:.4f}*x1 + {weights[1]:.4f}*x2 + {weights[2]:.4f}*x3 + {bias:.4f}")
print(f"True Equation: y = 2*x1 + 0.5*x2 - 1.5*x3 + 5 + noise")
print(f"Training MSE: {train_mse:.4f}")
print(f"Testing MSE: {test_mse:.4f}")
print(f"Training R²: {train_r2:.4f}")
print(f"Testing R²: {test_r2:.4f}")
```

```python
# Create a plot to compare actual vs predicted values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_test_pred, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Multiple Linear Regression: Actual vs Predicted')
plt.show()

# Visualize the coefficients
plt.figure(figsize=(10, 6))
plt.bar(['feature1', 'feature2', 'feature3'], weights)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.xlabel('Features')
plt.ylabel('Coefficient Value')
plt.title('Feature Coefficients in Multiple Linear Regression')
plt.show()
```
