```python
#A fim de explorar as variáveis disponíveis, faça o seguinte:

# A) Caracterize, em um curto parágrafo, o comportamento da variável ‘idade’; apresente um histograma de idade.
plt.figure(figsize=(8, 5))
sns.histplot(
    data=df,
    x='Idade',
    discrete=True,
    shrink=0.8
)
plt.title('Distribuição de Idade')
plt.xlabel('Idade')
plt.ylabel('Quantidade pessoas')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
print("A população pesquisada demonstra maior concentração em uma faixa etária mais jovem, entre 35 a 25 anos")
print("\n")
```

```python
# B) Apresente boxplots, lado a lado, para ‘Fut_Fin_WB’ entre indivíduos que trabalham e que não trabalham.
sns.boxplot(x="Trabalha", y="Fut_Fin_WB ", data=df)
plt.title("Distribuição da percepção de bem-estar financeiro futuro, por trabalham e não trabalham")
plt.xlabel("Trabalha")
plt.ylabel("Percepção")
plt.show()
```

```python
# C) Apresente um gráfico de barras para representar as frequências relativas ou absolutas da variável ‘cidade’. 
# c. Gráfico de barras para frequências da variável 'Cidade'
cidade_frequencias = df['Cidade'].value_counts()
plt.figure(figsize=(8, 6))
cidade_frequencias.plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title('Frequências Absolutas da Variável Cidade')
plt.xlabel('Cidade')
plt.ylabel('Frequência')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# D) Você acha que há evidência de que o conhecimento financeiro objetivo (variável ‘Fin_knowledge’) está relacionado com a autopercepção de bem-estar financeiro futuro (variável ‘Fut_Fin_WB’)?
sns.scatterplot(x="Fin_knowledge ", y="Fut_Fin_WB ", data=df)
plt.title("Relação entre conhecimento financeiro e  autopercepção")
plt.xlabel("conhecimento financeiro")
plt.ylabel("autopercepção")
plt.show()

print("Existe uma correlação positiva entre conhecimento financeiro e autopercepção de bem-estar financeiro futuro")
print("\n")
print("Isto é: quanto maior o conhecimento financeiro objetivo, melhor é a autopercepção de bem-estar financeiro futuro")
print("\n")
```

```python
#Adicionalmente Considere dois grupos
df_sp = df[df['Cidade'] == 'Sao Paulo']
df_rj = df[df['Cidade'] == 'Rio de Janeiro']
```

```python

# A) Caracterize, em um curto parágrafo, o comportamento da variável ‘idade’; apresente um histograma de idade.

# Histograma São Paulo
plt.subplot(1, 2, 1)
sns.histplot(df_sp['Idade'], bins=10, kde=False, color='seagreen', edgecolor='black')
plt.title('São Paulo')
plt.xlabel('Idade')
plt.ylabel('')
plt.grid(axis='y', linestyle='--', alpha=0.6)
# Histograma Rio de Janeiro
plt.subplot(1, 2, 2)
sns.histplot(df_rj['Idade'], bins=10, kde=False, color='indianred', edgecolor='black')
plt.title('Rio de Janeiro')
plt.xlabel('Idade')
plt.ylabel('')
plt.grid(axis='y', linestyle='--', alpha=0.6)
```

```python
# B) Apresente boxplots, lado a lado, para ‘Fut_Fin_WB’ entre indivíduos que trabalham e que não trabalham.
plt.subplot(1, 2, 1)
sns.boxplot(x="Trabalha", y="Fut_Fin_WB ", data=df_sp)
plt.title("Distribuição da percepção de bem-estar financeiro futuro, por trabalham e não trabalham em São Paulo")
plt.xlabel("Trabalha")
plt.ylabel("Percepção")
plt.show()
plt.subplot(1, 2, 2)
sns.boxplot(x="Trabalha", y="Fut_Fin_WB ", data=df_rj)
plt.title("Distribuição da percepção de bem-estar financeiro futuro, por trabalham e não trabalham em Rio de Janeiro")
plt.xlabel("Trabalha")
plt.ylabel("Percepção")
plt.show()
```

```python
# C) Apresente um gráfico de barras para representar as frequências relativas ou absolutas da variável ‘cidade’. 
# c. Gráfico de barras para frequências da variável 'Cidade'
df_cidades = df[df['Cidade'].isin(['Sao Paulo', 'Rio de Janeiro'])]
cidade_frequencias = df_cidades['Cidade'].value_counts()
plt.figure(figsize=(8, 6))
cidade_frequencias.plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title('Frequências Absolutas da Variável Cidade')
plt.xlabel('Cidade')
plt.ylabel('Frequência')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
# Calcular a média de idade para São Paulo
print(df_sp['Fut_Fin_WB '].describe())
print("\n")
# Calcular a média de idade para Rio de Janeiro
print(df_rj['Fut_Fin_WB '].describe())
```

```python
print("Apesar da média da percepção financeira do RJ ser menor do que de SP, a dispersão desse indicador é maior no RJ relativamente aos que não trabalham. \nEmbora as médias de idade sejam semelhantes entre as duas cidades, a distribuição etária revela diferenças importantes na composição das populações analisadas.")

```
